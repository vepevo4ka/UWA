Для мягкого дна, где $c_b < c$, критического отражения не происходит, поэтому оно не учитывается в формуле (4). |
| (4) | $\bar{H}(f) = \sum_p \bar{H}_p(f)e^{-j2\pi f \bar{\tau}_p}$ | $\bar{H}(f)$: Channel transfer function, $\bar{H}_p(f)$: Path transfer function, $\bar{\tau}_p$: Propagation delay, $\bar{\tau}_p = \frac{l_p}{c} - t_0$, where $t_0$ is a reference time, e.g., $t_0 = \frac{l_0}{c}$ for constant sound speed | Defines the overall **channel transfer function** of the multipath channel, which accounts for all propagation paths. | (5) | $H_p(f) \approx \frac{\Gamma_p}{\sqrt{\left(\frac{l_p}{l_0}\right)^k a(f)^{l_p - l_0}}} H_0(f)$ | $H_p(f)$: Path transfer function, $H_0(f)$: Reference transfer function, $\Gamma_p$: Reflection coefficient, $l_p, l_0$: Path lengths, $a(f)$: Absorption coefficient (see 6)| Approximation of the path transfer function. | Аппроксимация функции передачи пути. | (6) | $a(f)^{-(l_p - t_0)/2} \approx a_0^{-(l_p - t_0)/2}$ \\ $a(f)^{-\Delta l_p/2} = [1 + \epsilon(f)]^{-\Delta l_p/2} \approx 1 - \epsilon(f) \Delta l_p/2$ | $a(f)$: Absorption function, $a_0$: Absorption factor at reference frequency, $\Delta l_p = l_p - t_0$, $\epsilon(f)$: Small perturbation function | Approximates the **frequency-dependent absorption function**, assuming small variations. Достоверность приближения подтверждается числовыми значениями: при $\Delta l_p$ до 40 м значение $a(f)^{-\Delta l_p/2}$ изменяется лишь от 0.9 до 1.0, что позволяет заменить $a(f)$ на $a_0$. При этом $a_0$ можно выбирать для любой частоты в рабочей полосе, например, для центральной частоты $f_c$ или на границах полосы. |

| (8) | $\bar{h}_p = \frac{\Gamma_p}{\sqrt{\left(\frac{l_p}{l_0}\right)^k a_0^{l_p - t_0}}}$ | $\bar{h}_p$: Path gain, $\Gamma_p$: Reflection coefficient, $l_p, l_0$: Path lengths, $k$: Spreading exponent, $a_0$: Absorption function | Defines the **path gain**, incorporating the effects of reflection, path length, and absorption.
